# 🧠 YData Profiling Guide
This notebook demonstrates the full workflow for generating **data profiling reports** using `ydata-profiling`.

**Objectives:**
1. Install and setup YData Profiling
2. Load datasets
3. Generate profiling reports
4. Add metadata
5. Save reports
6. Automate multiple datasets

````python
# ----------------------------
# 📚 2. Import Libraries
# ----------------------------
import pandas as pd
from ydata_profiling import ProfileReport
from pathlib import Path
````

````python
# ----------------------------
# 📂 3. Load Dataset
# ----------------------------
# Load the customers dataset as an example
customers = pd.read_csv("data/olist_customers_dataset.csv")

# Preview the first 5 rows to understand the data
customers.head()
````

````python
# ----------------------------
# 🧾 4. Generate Profiling Report
# ----------------------------
# Create a basic profiling report for the dataset
profile = ProfileReport(
    df=customers,           # Pass the dataframe
    title="Customers Profiling Report",  
    explorative=True,        # Enables interactive visualizations
    config_file=""   # 👈 add this line, if you want it to work

)

# Save report to HTML
profile.to_file("profiling_reports/customers_profiling_report.html")

print("✅ Customers profiling report generated successfully!")
````

````python
# ----------------------------
# 🖋️ 5. Add Dataset Metadata (Optional)
# ----------------------------
# You can add dataset description, copyright, and URL info
profile.config.dataset.description = "Customer data profiling for Retail Sales Analysis project."
profile.config.dataset.copyright_holder = "Abe Temitope Moses"
profile.config.dataset.copyright_year = 2025
profile.config.dataset.url = "https://sites.google.com/view/abelytics"

# Save updated report
profile.to_file("../reports/customers_profiling_report.html")
print("✅ Customers profiling report with metadata updated successfully!")
````

````python
# ----------------------------
# 🔁 6. Automate Profiling for Multiple Tables
# ----------------------------
datasets = ["customers", "inventory", "order_details", "orders", "products", "returns", "warehouses"]

for name in datasets:
    df = pd.read_csv(f"../data/raw_data/{name}.csv")  # Load dataset
    profile = ProfileReport(df, title=f"{name.capitalize()} Profiling Report", explorative=True, config_file="")
    
    # Optional: Add metadata
    profile.config.dataset.description = f"{name.capitalize()} dataset profiling"
    profile.config.dataset.copyright_holder = "Abe Temitope Moses"
    profile.config.dataset.copyright_year = 2025
    profile.config.dataset.url = "https://sites.google.com/view/abelytics"
    
    profile.to_file(Path(f"../reports/data_profiling/{name}_profiling_report.html"))
    print(f"✅ {name.capitalize()} profiling report generated!")
````

# YData Profiling: `config_file` Explained

## What is `config_file`?

The `config_file` parameter in `ProfileReport()` allows you to **pass a YAML or JSON configuration file** that defines how the profiling report should behave.  
Think of it as a **settings file** for your profiling report.

---

## Purpose of `config_file`

- Customize report appearance (theme, colors, dark mode, navbar visibility)  
- Enable or disable sections (like correlations, missing values plots, duplicates)  
- Adjust computational settings (like sampling, lazy evaluation)  

Instead of manually setting all options in code, you can prepare a YAML file with your desired settings.

### Example YAML (`my_config.yaml`):

```yaml
title: "Custom Profiling Report"
dataset:
  description: "This is my dataset"
  copyright_holder: "Abe Temitope Moses"
  copyright_year: 2025
  url: "http://example.com"
html:
  use_local_assets: true
explorative: true


from ydata_profiling import ProfileReport

profile = ProfileReport(df=customers, config_file="my_config.yaml")
