# 📝 Matplotlib `text()` and `annotate()` – Complete Guide

## 📌 Overview

Both functions are used to **add text to a plot**:

* `text()` → for **simple text placement**
* `annotate()` → for **highlighting points (with arrows, boxes, etc.)**

---

# 🔤 `text()` Function

## 🧾 Syntax

```python
ax.text(x, y, s, fontsize=None, color=None, ha=None, va=None)
```

## 🔍 Parameters

* `x, y` → Position of the text
* `s` → Text string
* `fontsize` → Size of text
* `color` → Text color
* `ha` → Horizontal alignment (`'left'`, `'center'`, `'right'`)
* `va` → Vertical alignment (`'top'`, `'center'`, `'bottom'`)

---

## 📊 Example

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [10, 20, 15])

plt.text(2, 20, "Peak", fontsize=12)

plt.show()
```

## 🧠 What It Does

* Places the text **"Peak"** at coordinate `(2, 20)`
* No arrow or pointer — just plain text

---

# 🏷️ `annotate()` Function

## 🧾 Syntax

```python
ax.annotate(text, xy, xytext=None, arrowprops=None)
```

## 🔍 Parameters

* `text` → Annotation text
* `xy` → Point being highlighted
* `xytext` → Position of the text
* `arrowprops` → Dictionary for arrow styling

---

## 📊 Example

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [10, 20, 15]

plt.plot(x, y)

plt.annotate(
    "Highest Point",
    xy=(2, 20),
    xytext=(2.5, 25),
    arrowprops=dict(facecolor='black')
)

plt.show()
```

## 🧠 What It Does

* Highlights point `(2, 20)`
* Places text above it
* Draws an arrow from text → point

---

# 🚀 Advanced `annotate()` Examples

## 🎯 Styled Annotation (Box + Arrow)

```python
plt.annotate(
    "Peak Value",
    xy=(2, 20),
    xytext=(2.5, 25),
    arrowprops=dict(
        arrowstyle="->",
        color="black"
    ),
    bbox=dict(
        boxstyle="round,pad=0.3",
        fc="yellow",
        ec="black"
    )
)
```

---

## 🎯 Curved Arrow

```python
plt.annotate(
    "Curved Arrow",
    xy=(2, 20),
    xytext=(3, 25),
    arrowprops=dict(
        arrowstyle="->",
        connectionstyle="arc3,rad=0.3"
    )
)
```

---

## 🎯 Automatically Annotate Maximum Point

```python
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 25, 18, 30, 22])

plt.plot(x, y)

max_x = x[np.argmax(y)]
max_y = max(y)

plt.annotate(
    f"Max: {max_y}",
    xy=(max_x, max_y),
    xytext=(max_x, max_y + 5),
    arrowprops=dict(arrowstyle="->")
)
```

---

## 🎯 Annotate Multiple Points

```python
for i in range(len(x)):
    plt.annotate(
        y[i],
        (x[i], y[i]),
        textcoords="offset points",
        xytext=(0, 5),
        ha='center'
    )
```

# ⚖️ Key Differences

| Feature         | `text()` 📝   | `annotate()` 🏷️    |
| --------------- | ------------- | ------------------- |
| Basic text      | ✅ Yes         | ✅ Yes               |
| Arrow support   | ❌ No          | ✅ Yes               |
| Highlight data  | ❌ No          | ✅ Yes               |
| Styling (boxes) | ❌ No          | ✅ Yes               |
| Use case        | Labels, notes | Highlighting points |

---

# 💡 Best Practices

* Use `text()` for **simple labels or notes**
* Use `annotate()` to **draw attention to key data points**
* Avoid too many annotations to keep charts clean
* Use `bbox` in `annotate()` for better readability
* Use `textcoords="offset points"` to prevent overlap

---

# ✅ Quick Summary

* `text()` = simple text
* `annotate()` = text + arrow + highlighting
