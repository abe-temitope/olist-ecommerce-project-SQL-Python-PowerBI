# 🔁 Ways to Iterate Over a DataFrame in Pandas

This document explains the main methods used to loop through a DataFrame in **:contentReference[oaicite:0]{index=0}**, with clear syntax and examples.

---

# 📊 1. `iterrows()` (Row-wise iteration)

## 🧾 Syntax
```python
for index, row in df.iterrows():
    ...
````

## 🧠 Description

* Iterates through a DataFrame **row by row**
* Each row is returned as a **Pandas Series**

## 📌 Example

```python id="ex1"
for index, row in df.iterrows():
    print(index, row["sales"], row["product"])
```

## 📊 Use case

* Adding labels to charts
* Simple row-based logic
* Small to medium datasets

---

# ⚡ 2. `itertuples()` (Faster row-wise iteration)

## 🧾 Syntax

```python
for row in df.itertuples():
    ...
```

## 🧠 Description

* Faster alternative to `iterrows()`
* Returns each row as a **named tuple**
* More memory efficient

## 📌 Example

```python id="ex2"
for row in df.itertuples():
    print(row.Index, row.sales, row.product)
```

## 📊 Use case

* Large datasets
* Performance-sensitive operations
* Production-level loops

---

# 🧮 3. `apply()` (Function-based iteration)

## 🧾 Syntax

```python
df.apply(function, axis=1)
```

## 🧠 Description

* Applies a function to each row or column
* `axis=1` → row-wise
* `axis=0` → column-wise

## 📌 Example (Row-wise)

```python id="ex3"
df["profit_flag"] = df.apply(
    lambda row: "High" if row["profit"] > 100 else "Low",
    axis=1
)
```

## 📊 Use case

* Feature engineering
* Custom transformations
* Conditional logic

---

# 📦 4. Column Iteration (`df.columns`)

## 🧾 Syntax

```python
for col in df.columns:
    ...
```

## 🧠 Description

* Iterates through **column names**
* Each column can be accessed using `df[col]`

## 📌 Example

```python id="ex4"
for col in df.columns:
    print(col, df[col].mean())
```

## 📊 Use case

* Feature analysis
* Column profiling
* Data inspection

---

# 🔀 5. `zip()` (Multi-column iteration)

## 🧾 Syntax

```python
for a, b in zip(df["col1"], df["col2"]):
    ...
```

## 🧠 Description

* Iterates over multiple columns simultaneously
* Pairs values row-by-row

## 📌 Example

```python id="ex5"
for revenue, cost in zip(df["revenue"], df["cost"]):
    print(revenue - cost)
```

## 📊 Use case

* Comparing two columns
* Row-wise calculations without apply()

---

# 🔥 Summary Table

| Method         | Type         | Speed  | Best Use Case             |
| -------------- | ------------ | ------ | ------------------------- |
| `iterrows()`   | Row-wise     | Slow   | Simple loops, annotations |
| `itertuples()` | Row-wise     | Fast   | Large datasets            |
| `apply()`      | Row/Column   | Medium | Feature engineering       |
| `df.columns`   | Column-wise  | Fast   | Feature analysis          |
| `zip()`        | Multi-column | Fast   | Column comparisons        |

---

# 🧠 Key Insight

* Rows → `iterrows()` / `itertuples()`
* Columns → `df.columns`
* Multiple columns → `zip()`
* Best performance → `apply()` or vectorized operations

---
