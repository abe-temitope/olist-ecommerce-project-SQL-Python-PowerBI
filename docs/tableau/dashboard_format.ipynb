# 📊 Olist E-Commerce Analytics Dashboard Design System

## 🎯 Purpose
This document defines the **standard design system** used for building all dashboards in the Olist E-Commerce Analytics project using :contentReference[oaicite:0]{index=0}.  
It ensures consistency, readability, and professional-level presentation suitable for portfolio and client delivery.

---

# 🧠 Design Philosophy
- Business-first, not decoration-first
- Clarity over complexity
- Insights over visuals
- Minimal but powerful design

---

# 🖋️ 1. TYPOGRAPHY SYSTEM

## 🔤 Primary Font
- **Segoe UI**
- Alternatives: Inter, Helvetica, Arial

---

## 📏 Font Hierarchy

| Element | Size | Weight | Style |
|--------|------|--------|------|
| Dashboard Title | 22–26px | Bold | Sentence case |
| Section Headers | 14–16px | Semi-bold | Clean |
| KPI Values | 28–42px | Bold | Prominent |
| KPI Labels | 10–12px | Regular | Uppercase optional |
| Chart Titles | 12–14px | Semi-bold | Clear naming |
| Axis Labels | 9–11px | Regular | Minimal |
| Tooltips | 10–11px | Regular | Readable |

---

## 🧠 Typography Rules
- Use **bold only for emphasis**
- Avoid multiple font styles in one view
- Maintain consistency across all dashboards
- Keep labels short and business-oriented

---

# 🎨 2. COLOR SYSTEM

## 🌐 Primary Palette

| Purpose                 | Color Meaning             | HEX       | Usage                                     |
| ----------------------- | ------------------------- | --------- | ----------------------------------------- |
| Background / Base Theme | Neutral Surface           | `#EDE9E6` | Dashboard background, containers, cards   |
| Positive Performance    | Growth / Profit           | `#C9996B` | Revenue, profit, upward trends            |
| Negative Performance    | Risk / Loss / Issues      | `#5C4F4A` | Loss, cancellations, delays, warnings     |
| Secondary Context       | Support / Neutral Insight | `#5C766D` | Filters, secondary charts, maps, grouping |

---

## 🎯 Color Usage Rules
- One dashboard = one dominant theme (blue preferred)
- Green = positive business outcomes only
- Red = losses, delays, cancellations only
- Avoid unnecessary color variety
- Never use decorative colors without meaning

---

# 📐 3. DASHBOARD LAYOUT STRUCTURE

## 🧱 Standard Layout Framework

```

---

## HEADER (Title + Filters)

## KPI ROW (4–6 KPIs)

## MAIN VISUAL AREA (Charts)

## INSIGHTS PANEL (Text)

## FILTER PANEL (Sidebar)

```

---

## 📊 Section Breakdown

### 🔵 Header
- Dashboard title
- Date filter
- Optional global filters

---

### 🟡 KPI Row (Top Priority Section)
Must include:
- Total Revenue
- Total Orders
- Average Order Value
- Delivery Rate
- Review Score (optional)

👉 KPI cards must be visually dominant

---

### 📈 Main Visual Area
Include 2–4 key visuals only:
- Revenue trend (line chart)
- Category performance (bar chart)
- Geographic distribution (map)
- Customer behavior analysis

---

### 🧠 Insights Panel
- 3–5 key business insights
- Written in simple business language
- No technical jargon

Example:
- “Technology category generates highest revenue contribution”
- “Delayed deliveries correlate with lower review scores”

---

### 🎛️ Filter Panel
- Date range
- Category
- State
- Order status

Keep filters vertical and clean

---

# 📊 4. KPI DESIGN STANDARDS

## Layout Rules
- Large number (28–42px)
- Label below number (10–12px)
- Center or left aligned
- Minimal background clutter

---

## KPI Example Format

```

💰 2.4M
Total Revenue

```

---

# 📈 5. VISUALIZATION STANDARDS

## Recommended Charts

### 📊 Bar Chart
- Best for comparisons
- Example: Sales by category

---

### 📈 Line Chart
- Best for trends
- Example: Revenue over time

---

### 🌍 Map Visualization
- Geographic insights
- Example: Customer distribution

---

## ❌ Avoid
- Pie charts (except very simple breakdowns)
- 3D charts
- Overcrowded dashboards
- Too many colors

---

# 🧠 6. DATA INTERPRETATION PRINCIPLES

Every visual must answer:

### 1. What happened?
- Revenue increased/decreased

### 2. Where did it happen?
- Category, region, customer segment

### 3. Why did it happen?
- Delivery delays, product mix, customer behavior

---

# ⚙️ 7. INTERACTIVITY DESIGN

## Required Elements
- Filters (global and local)
- Hover tooltips
- Click-to-filter actions

---

## Advanced Elements (Optional but Powerful)
- Parameter controls
- Set-based segmentation
- Dashboard navigation buttons
- Highlight actions

---

# 🧩 8. DATA MODEL AWARENESS

This dashboard system is built on a star-schema model:

### Fact Tables
- Orders
- Order Items
- Payments
- Reviews

### Dimension Tables
- Customers
- Sellers
- Products
- Category Translation
- Geolocation (role-based dimensions)

---

# 🌍 9. GEOGRAPHY DESIGN RULE

- Use separate roles:
  - Customer Geography
  - Seller Geography

- Always ensure:
  - Zip code prefix is the join key
  - Cleaned geolocation table is used

---

# 🚀 10. PROFESSIONAL STANDARDS

A dashboard is considered “portfolio-ready” if it has:

✔ Clean layout  
✔ Consistent typography  
✔ Minimal color usage  
✔ KPI-driven structure  
✔ Business insights section  
✔ Interactive filters  
✔ Clear storytelling flow
