# 🎨 📊 TABLEAU DASHBOARD STYLE

---

# 🖋️ 1. FONT SYSTEM (VERY IMPORTANT)

## ✅ Primary Font

* **Segoe UI** (best default for business dashboards)
* Alternative: Inter / Helvetica

---

## 📏 Font hierarchy

| Element         | Size    | Weight    | Style              |
| --------------- | ------- | --------- | ------------------ |
| Dashboard Title | 20–24px | Bold      | Uppercase optional |
| Section Headers | 14–16px | Semi-bold | Sentence case      |
| KPI Numbers     | 28–40px | Bold      | Centered or left   |
| KPI Labels      | 10–12px | Regular   | Uppercase optional |
| Chart Titles    | 12–14px | Semi-bold | Clean              |
| Axis Labels     | 9–11px  | Regular   | Minimal            |
| Tooltips        | 10–11px | Regular   | Auto               |

---

## 🧠 Rule:

👉 Big numbers = bold
👉 Everything else = minimal weight

---

# 🎨 2. COLOR SYSTEM (CLEAN + CORPORATE)

## 🌐 Primary Palette

### 🔵 Primary Blue (Main brand color)

* #2F80ED

### 🟢 Profit / Positive

* #27AE60

### 🔴 Loss / Negative

* #EB5757

### 🟡 Warning / Neutral highlight

* #F2C94C

### ⚫ Text (dark UI)

* #2C2C2C

### ⚪ Background

* #F7F9FC

---

## 🧠 Color Rules (VERY IMPORTANT)

✔ Use **blue as default theme color**
✔ Use green/red ONLY for meaning
❌ Don’t use random colors per chart
❌ Avoid rainbow dashboards

---

# 📐 3. LAYOUT STRUCTURE (PRO LEVEL)

## 🧱 Standard dashboard layout:

```
[ HEADER ]
Title + Date Filter

[ KPI ROW ]
Revenue | Orders | AOV | Delivery Rate

[ MAIN VISUALS ]
Chart 1 | Chart 2 | Map / Trend

[ INSIGHTS PANEL ]
Key business insights (text)

[ FILTER PANEL ]
Category | Date | State | Status
```

---

## 📊 Spacing rules:

* Minimum padding: 10–20px
* Keep white space (do NOT overfill)
* Align everything to grid

---

# 📊 4. CHART STYLE RULES

## 📈 Line chart

* Use for time trends
* Example: Revenue over time

## 📊 Bar chart

* Use for comparison
* Example: Category sales

## 🌍 Map

* Use for geography analysis
* Example: customer distribution

## ❌ Avoid:

* Pie charts (unless very small categories)
* 3D charts
* Overdecorated visuals

---

# 📌 5. KPI CARD STYLE

## Design:

* Background: white or light grey
* Border: subtle shadow or none
* Value: BIG (28–40px)
* Label: small (10–12px)

---

## Example layout:

```
💰 2.4M
Total Revenue
```

---

# 🎯 6. INTERACTIVITY STYLE

Use only:

* Filters (left panel)
* Hover tooltips
* Click actions (drill-down)

---

## Optional advanced:

* Parameter controls
* Set-based filtering
* Dashboard actions (click → update charts)

---

# 🧠 7. TEXT STYLE (VERY IMPORTANT)

## Rule:

👉 Always write like a business report

❌ Bad:
“SUM(Sales by Category)”

✔ Good:
“Revenue by Category”

---

# 🚀 8. PROFESSIONAL DASHBOARD FEEL

Your dashboard should feel:

✔ Clean
✔ Minimal
✔ Decision-focused
✔ Not decorative

---

# 💡 GOLDEN RULE (MOST IMPORTANT)

> “If removing a chart makes the dashboard clearer, remove it.”
