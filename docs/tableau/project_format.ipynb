# 📊 Olist E-Commerce Analytics Project

## 📌 Project Overview
This project analyzes the Olist Brazilian e-commerce dataset to uncover insights on sales performance, customer behavior, delivery efficiency, and product trends. The goal is to transform raw transactional data into actionable business intelligence using SQL, Python, and Tableau.

---

## 🎯 Business Objectives
- Understand overall sales and revenue performance
- Identify top-performing products and categories
- Analyze customer purchasing behavior and retention patterns
- Evaluate delivery performance and order fulfillment efficiency
- Map geographic distribution of customers and sellers

---

## 🧰 Tools & Technologies Used
- SQL (MySQL Workbench) – Data extraction and transformation
- Python (pandas, numpy) – Data cleaning and feature engineering
- Tableau – Data visualization and dashboard creation
- Excel/CSV – Data storage and export format

---

## 🗂️ Dataset Description
The dataset contains 9 related tables:

- Orders (core transactional table)
- Order Items (product-level transactions)
- Customers
- Sellers
- Products
- Payments
- Reviews
- Geolocation
- Product Category Translation

---

## 🏗️ Data Model Design
The project follows a **star schema structure**:

- **Fact Table:**
  - Orders
  - Order Items
  - Payments
  - Reviews

- **Dimension Tables:**
  - Customers
  - Sellers
  - Products
  - Category Translation
  - Geolocation (role-playing dimension)

---

## 🧹 Data Cleaning Process
- Handled missing values using appropriate imputation techniques
- Removed duplicate records
- Standardized column naming conventions
- Converted timestamp fields to datetime format
- Aggregated geolocation data to zip-code level (deduplication)
- Created derived features such as delivery time and order status grouping

---

## 🔗 Data Relationships
- Orders → Customers (customer_id)
- Orders → Order Items (order_id)
- Orders → Payments (order_id)
- Orders → Reviews (order_id)
- Order Items → Products (product_id)
- Products → Category Translation (product_category_name)
- Customers → Geolocation (zip_code_prefix)
- Sellers → Geolocation (zip_code_prefix)

## 📌 Key Insights
- A small percentage of products contribute to a large portion of revenue
- Delivery delays negatively impact customer review scores
- Certain regions show higher order density and repeat purchases
- Payment methods vary significantly across customer segments
