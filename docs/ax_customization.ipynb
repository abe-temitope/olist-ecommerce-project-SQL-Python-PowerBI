## 🧠 Core Idea

```text
Seaborn → creates the plot
Matplotlib (ax) → customizes the plot
````

👉 You should ALWAYS think this way.

---

## 🔌 Setup

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

---

# 🔥 1. BASIC STRUCTURE (BEST PRACTICE)

```python
fig, ax = plt.subplots(figsize=(8, 5))

sns.barplot(data=df, x="month", y="revenue", ax=ax)

plt.show()
```

---

# 🔥 2. MOST IMPORTANT AX FUNCTIONS

---

## 🏷️ Titles & Labels

```python
ax.set_title("Monthly Revenue", fontsize=14)
ax.set_xlabel("Month", fontsize=12)
ax.set_ylabel("Revenue (USD)", fontsize=12)
```

---

## 🔢 Axis Limits

```python
ax.set_ylim(0, 20000)
ax.set_xlim(0, 10)
```

---

## 🎯 Ticks Control

```python
ax.set_xticks([0, 1, 2])
ax.set_yticks([0, 5000, 10000])
```

---

## 🔁 Tick Labels

```python
ax.set_xticklabels(["Jan", "Feb", "Mar"], rotation=45)
```

---

# 🔥 3. FORMAT NUMBERS (VERY IMPORTANT)

---

## 💰 Currency Formatting

```python
ax.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, _: f"${x:,.0f}")
)
```

---

## 📊 Percentage

```python
ax.yaxis.set_major_formatter(ticker.PercentFormatter())
```

---

## 🔢 Thousands (K, M)

```python
ax.yaxis.set_major_formatter(ticker.EngFormatter())
```

---

# 🔥 4. CONTROL TICK DENSITY

```python
ax.yaxis.set_major_locator(ticker.MaxNLocator(nbins=5))
```

---

# 🔥 5. LEGEND CONTROL

```python
ax.legend(title="Category", loc="upper right")
```

---

## ❌ Remove legend

```python
ax.legend().remove()
```

---

# 🔥 6. GRID (VERY IMPORTANT FOR DASHBOARDS)

```python
ax.grid(True)
```

---

## Clean style

```python
ax.grid(axis="y", linestyle="--", alpha=0.7)
```

---

# 🔥 7. REMOVE SPINES (MAKE IT CLEAN)

```python
sns.despine()
```

OR

```python
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
```

---

# 🔥 8. BAR LABELS (VERY IMPORTANT 🔥)

```python
for container in ax.containers:
    ax.bar_label(container, fmt="%.0f")
```

---

## Currency Labels

```python
for container in ax.containers:
    ax.bar_label(container, fmt="$%.0f")
```

---

# 🔥 9. ANNOTATION (ADVANCED)

```python
ax.annotate(
    "Peak",
    xy=("Mar", 12000),
    xytext=("Feb", 15000),
    arrowprops=dict(arrowstyle="->")
)
```

---

# 🔥 10. ROTATE LABELS

```python
plt.xticks(rotation=45)
```

---

# 🔥 11. SORT DATA BEFORE PLOTTING

```python
df = df.sort_values(by="revenue", ascending=False)
```

---

# 🔥 12. MULTIPLE PLOTS (SUBPLOTS)

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

sns.barplot(data=df, x="month", y="revenue", ax=ax[0])
sns.lineplot(data=df, x="month", y="revenue", ax=ax[1])
```

---

# 🔥 13. STYLE (SEABORN)

```python
sns.set_style("whitegrid")
sns.set_context("talk")
```

---

# 🔥 14. COLOR CONTROL

```python
sns.barplot(data=df, x="month", y="revenue", palette="Blues")
```

---

# 🔥 15. FULL PROFESSIONAL EXAMPLE

```python
fig, ax = plt.subplots(figsize=(8, 5))

sns.barplot(data=df, x="month", y="revenue", ax=ax, palette="Blues")

# Titles
ax.set_title("Monthly Revenue", fontsize=14)
ax.set_xlabel("Month")
ax.set_ylabel("Revenue (USD)")

# Format currency
ax.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, _: f"${x:,.0f}")
)

# Reduce ticks
ax.yaxis.set_major_locator(ticker.MaxNLocator(5))

# Add labels
for container in ax.containers:
    ax.bar_label(container, fmt="$%.0f")

# Clean look
ax.grid(axis="y", linestyle="--", alpha=0.7)
sns.despine()

plt.xticks(rotation=45)

plt.show()
```

---

# 🧠 FINAL MINDSET

```text
Plot → Customize → Clean → Label → Format
```

---

# 🚀 WHAT MAKES YOUR CHART PROFESSIONAL

✅ Clean axis (no clutter)
✅ Proper number formatting ($, %, K, M)
✅ Labels on bars
✅ Light grid
✅ No unnecessary borders
✅ Sorted data

---

# 🎯 GOLDEN RULE

```text
If your chart is hard to read → it's wrong
If it's simple and clear → it's professional
```

```

---

If you want next 🔥  
I can show you how to **convert these charts into a full Streamlit dashboard (Power BI style with filters, KPIs, cards, interactivity)** — that’s where your skill becomes monetizable 💰
```
