# 🥧 Pie Chart using Your Style (fig, ax)

Pie charts are created using **:contentReference[oaicite:0]{index=0}**, and the chart is built inside the `ax` object.

---

## 🧾 Basic Syntax (Your Style)

```python
fig, ax = plt.subplots()

ax.pie(values, labels=labels)

plt.show()
````

---

# 🧠 Breakdown (Your Style Thinking)

## 📊 1. `fig`

👉 The full canvas (entire figure)

* Think of it as the dashboard page
* It holds everything in the visualization

---

## 📍 2. `ax`

👉 The chart area

* This is where the pie chart is drawn
* All visual elements are created inside `ax`

---

## 🥧 3. `ax.pie()`

👉 This function creates the pie chart

---

# 📦 Full Syntax Explained

```python
ax.pie(
    x,              # values (sizes of slices)
    labels=None,    # category names
    colors=None,    # slice colors
    autopct=None,   # percentage display
    explode=None,   # highlight slices
    startangle=None # rotation of chart
)
```

---

# 🔍 Example (Your Style)

```python
import matplotlib.pyplot as plt

labels = ["Electronics", "Fashion", "Home"]
values = [40, 35, 25]

fig, ax = plt.subplots()

ax.pie(
    values,
    labels=labels
)

plt.show()
```

---

# 🎨 Improved Example (Professional Style)

```python
colors = ["#FF6A00", "#2E86AB", "#1B998B"]
explode = [0.1, 0, 0]

fig, ax = plt.subplots()

ax.pie(
    values,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%",
    explode=explode,
    startangle=90
)

ax.set_title("Sales Distribution by Category")

plt.show()
```

---

# 🧠 Parameter Explanation

## 📊 `values`

* Defines the size of each slice

## 🏷️ `labels`

* Names shown on the chart

## 🎨 `colors`

* Custom colors for slices

## 📈 `autopct`

* Displays percentage values on slices

## 🔥 `explode`

* Pulls out a slice for emphasis

## 🧭 `startangle`

* Rotates the chart for better layout

---

# 📊 Real Project Use (Olist Style)

Pie charts are commonly used for:

* Revenue share by category
* Order distribution by status
* Payment method breakdown

---

# 🚀 Final Mental Model

```text
fig → full figure (dashboard)
ax  → chart area
ax.pie() → creates the pie chart inside ax
```

---

# 💡 Key Takeaway

```python
fig, ax = plt.subplots()
ax.pie(...)
```

* `fig` = container
* `ax` = drawing area
* `ax.pie()` = creates the visualization

---
