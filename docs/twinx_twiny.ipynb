# 📊 Matplotlib Dual Axis Guide: `twinx()` & `twiny()`

---

# 🧠 CORE IDEA

```text
twinx() → second Y-axis (right)
twiny() → second X-axis (top)
````

👉 Both create a **shared axis + independent axis**

---

# 🔥 1. `twinx()` — FULL BREAKDOWN

```python
ax2 = ax1.twinx()
```

### 📌 What it does:

* Shares **X-axis**
* Creates a **new Y-axis (right side)**

---

## ✅ ASSOCIATED FUNCTIONS (FOR `twinx()`)

---

## 🏷️ 1. `set_ylabel()`

```python
ax1.set_ylabel("Revenue ($)")
ax2.set_ylabel("Conversion (%)")
```

---

## 🔢 2. `set_ylim()`

```python
ax1.set_ylim(0, 5000)
ax2.set_ylim(0, 10)
```

---

## 🎯 3. `tick_params()`

```python
ax1.tick_params(axis='y')
ax2.tick_params(axis='y')
```

---

## 🔁 4. `yaxis.set_major_formatter()`

```python
import matplotlib.ticker as ticker

ax1.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, _: f"${x:,.0f}")
)

ax2.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, _: f"{x}%")
)
```

---

## 📏 5. `yaxis.set_major_locator()`

```python
ax1.yaxis.set_major_locator(ticker.MaxNLocator(5))
ax2.yaxis.set_major_locator(ticker.MaxNLocator(4))
```

---

## 🧱 6. `spines`

```python
ax2.spines["right"].set_visible(True)
ax1.spines["left"].set_visible(True)
```

Optional cleanup:

```python
ax1.spines["top"].set_visible(False)
```

---

## 📊 7. `legend()` (IMPORTANT ⚠️)

```python
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax1.legend(lines1 + lines2, labels1 + labels2)
```

---

## 📌 8. `plot()`

```python
ax1.plot(x, revenue, label="Revenue")
ax2.plot(x, conversion, label="Conversion")
```

---

## 📉 9. `grid()`

```python
ax1.grid(True)
```

---

## ✍️ 10. `annotate()`

```python
ax1.annotate("Peak", xy=(2, 2000))
```

---

# 🔥 FULL `twinx()` EXAMPLE

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

x = ["Jan", "Feb", "Mar"]
revenue = [1000, 1500, 2000]
conversion = [2.5, 3.0, 4.2]

fig, ax1 = plt.subplots()

# First axis
ax1.plot(x, revenue, label="Revenue")
ax1.set_ylabel("Revenue ($)")
ax1.set_ylim(0, 2500)

ax1.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, _: f"${x:,.0f}")
)

# Second axis
ax2 = ax1.twinx()
ax2.plot(x, conversion, label="Conversion")
ax2.set_ylabel("Conversion (%)")
ax2.set_ylim(0, 5)

# Combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2)

# Grid
ax1.grid(True)

plt.show()
```

---

# 🔥 2. `twiny()` — FULL BREAKDOWN

```python
ax2 = ax1.twiny()
```

### 📌 What it does:

* Shares **Y-axis**
* Creates a **new X-axis (top)**

---

## ✅ ASSOCIATED FUNCTIONS (FOR `twiny()`)

---

## 🏷️ 1. `set_xlabel()`

```python
ax1.set_xlabel("Days")
ax2.set_xlabel("Weeks")
```

---

## 🔢 2. `set_xlim()`

```python
ax1.set_xlim(0, 10)
ax2.set_xlim(0, 2)
```

---

## 🎯 3. `tick_params()`

```python
ax1.tick_params(axis='x')
ax2.tick_params(axis='x')
```

---

## 🔁 4. `xaxis.set_major_formatter()`

```python
ax2.xaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, _: f"{x}w")
)
```

---

## 📏 5. `xaxis.set_major_locator()`

```python
ax2.xaxis.set_major_locator(ticker.MaxNLocator(5))
```

---

## 🧱 6. `spines`

```python
ax2.spines["top"].set_visible(True)
```

---

## 📊 7. `plot()`

```python
ax1.plot(x, y)
```

---

# 🔥 FULL `twiny()` EXAMPLE

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

x_days = [1,2,3,4,5]
y = [10,20,30,40,50]

fig, ax1 = plt.subplots()

# Bottom axis
ax1.plot(x_days, y)
ax1.set_xlabel("Days")

# Top axis
ax2 = ax1.twiny()
ax2.set_xlim(ax1.get_xlim())

ax2.set_xlabel("Weeks")
ax2.xaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, _: f"{x/7:.1f}")
)

plt.show()
```

---

# 🧠 DIFFERENCE SUMMARY

| Feature     | `twinx()`    | `twiny()`       |
| ----------- | ------------ | --------------- |
| Shared axis | X            | Y               |
| New axis    | Y (right)    | X (top)         |
| Use case    | Dual metrics | Dual scale/time |

---

# ⚠️ CRITICAL RULES

```text
1. Always label both axes clearly
2. Keep scales meaningful
3. Avoid visual clutter
4. Combine legends properly
```

---

# 🎯 FINAL MINDSET

```text
twinx/twiny are not just functions —
they are tools for multi-dimensional storytelling
```
