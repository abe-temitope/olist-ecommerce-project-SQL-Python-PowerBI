# рџ§  1. What is an Artist?

An **Artist** is a base object in Matplotlib that represents anything drawn on a figure.

## рџЋЇ Two main categories:

### рџ§± Primitive Artists

Basic visual elements:

* Line2D (lines)
* Text (labels)
* Patch (shapes)
* Collection (groups of points)

### рџ“¦ Container Artists

Group multiple artists:

* BarContainer
* ErrorbarContainer
* Legend
