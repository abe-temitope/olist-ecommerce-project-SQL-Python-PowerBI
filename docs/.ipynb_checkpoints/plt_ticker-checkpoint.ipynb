# 📘 Matplotlib Ticker Cheat Sheet

---

## 🧠 What is matplotlib.ticker?

`matplotlib.ticker` is used to **control axis ticks and labels**.

👉 It handles:
- Tick positions → **Locators**
- Tick formatting → **Formatters**

---

## 🔌 Import

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
````

---

# 🔥 1. LOCATORS (Control Tick Positions)

---

## 📍 Common Locators

### 1. AutoLocator (default)

```python
ax.yaxis.set_major_locator(ticker.AutoLocator())
```

👉 Automatically chooses tick positions

---

### 2. MaxNLocator (most used 🔥)

```python
ax.yaxis.set_major_locator(ticker.MaxNLocator(nbins=5))
```

👉 Limits number of ticks

---

### 3. MultipleLocator

```python
ax.yaxis.set_major_locator(ticker.MultipleLocator(1000))
```

👉 Tick every 1000 units

---

### 4. FixedLocator

```python
ax.yaxis.set_major_locator(ticker.FixedLocator([0, 500, 1000]))
```

👉 Manual tick positions

---

### 5. LogLocator

```python
ax.yaxis.set_major_locator(ticker.LogLocator())
```

👉 For logarithmic scale

---

### 6. NullLocator

```python
ax.yaxis.set_major_locator(ticker.NullLocator())
```

👉 Removes ticks

---

# 🔥 2. FORMATTERS (Control Tick Labels)

---

## 📊 Common Formatters

### 1. ScalarFormatter (default)

```python
ax.yaxis.set_major_formatter(ticker.ScalarFormatter())
```

👉 Standard numbers (can disable scientific notation)

---

### 2. FuncFormatter (VERY IMPORTANT 🔥)

```python
formatter = ticker.FuncFormatter(lambda x, pos: f"${x:,.0f}")
ax.yaxis.set_major_formatter(formatter)
```

👉 Fully customizable

---

### 3. StrMethodFormatter

```python
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
```

👉 Uses Python format strings

---

### 4. FormatStrFormatter

```python
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter("%.2f"))
```

👉 Old-style formatting

---

### 5. PercentFormatter

```python
ax.yaxis.set_major_formatter(ticker.PercentFormatter())
```

👉 Converts to percentage

---

### 6. EngFormatter

```python
ax.yaxis.set_major_formatter(ticker.EngFormatter())
```

👉 Engineering format (K, M, G)

---

### 7. LogFormatter

```python
ax.yaxis.set_major_formatter(ticker.LogFormatter())
```

👉 For log scale

---

### 8. NullFormatter

```python
ax.yaxis.set_major_formatter(ticker.NullFormatter())
```

👉 Hides labels

---

# 🔥 3. FULL EXAMPLE

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

x = ["Jan", "Feb", "Mar"]
y = [1000, 5000, 12000]

fig, ax = plt.subplots()
ax.bar(x, y)

# Control ticks
ax.yaxis.set_major_locator(ticker.MaxNLocator(nbins=4))

# Format labels (currency)
ax.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, pos: f"${x:,.0f}")
)

plt.show()
```

---

# 🔥 4. WHEN TO USE WHAT

| Situation            | Use              |
| -------------------- | ---------------- |
| Too many ticks       | MaxNLocator      |
| Fixed intervals      | MultipleLocator  |
| Custom labels        | FuncFormatter    |
| Currency             | FuncFormatter    |
| Percentage           | PercentFormatter |
| Large numbers (K, M) | EngFormatter     |

---

# ⚠️ COMMON ERROR

### ❌ Error

```python
ax.ticklabel_format(style='plain')
```

👉 Fails if using `FuncFormatter`

---

### ✅ Fix

Use `ScalarFormatter`:

```python
ax.yaxis.set_major_formatter(ticker.ScalarFormatter())
```

OR use `FuncFormatter` instead

---

# 🧠 IMPORTANT CONCEPT

```text
Locator = WHERE ticks appear
Formatter = HOW ticks look
```

---

# 🚀 PRO TIPS

✅ Use `FuncFormatter` for dashboards

```python
lambda x, _: f"${x:,.0f}"
```

✅ Combine locator + formatter for clean visuals

```python
ax.yaxis.set_major_locator(ticker.MaxNLocator(5))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f"{x:,.0f}"))
```

---

# 🎯 FINAL MINDSET

```text
Messy axis → bad visualization
Clean ticks → professional dashboard
```

```

---

If you want next 🔥  
I can give you a **Matplotlib + Seaborn + Ax customization cheat sheet (complete visualization mastery)** — that’s what makes your charts look like Power BI level.
```
