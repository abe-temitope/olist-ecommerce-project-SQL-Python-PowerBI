# 🔹 What is `deep-translator`?

## 🧠 Simple definition:
`deep-translator` is a Python library used to **translate text from one language to another** using multiple translation engines (like Google Translate).

👉 It helps convert text data into a target language for analysis.

````

---

# 🔧 Importing the module

```python
from deep_translator import GoogleTranslator
```

---

# 🌐 Basic Syntax

```python
GoogleTranslator(source='auto', target='en').translate(text)
```

---

# 🧠 Parameters Explained

## 🔹 `source`

* Language of input text
* `'auto'` = automatically detect language

```python
source='auto'
```

---

## 🔹 `target`

* Language you want to translate into

```python
target='en'   # English
```

---

## 🔹 `translate(text)`

* The actual function that translates text

```python
translate("produto ótimo")
```

---

# 📌 Simple Example

```python
from deep_translator import GoogleTranslator

translated = GoogleTranslator(
    source='auto',
    target='en'
).translate("produto ótimo")

print(translated)
```

👉 Output:

```
great product
```

---

# 📊 Using with Pandas (Single Column)

## ✔ Basic approach

```python
df["translated_text"] = df["text_column"].apply(
    lambda x: GoogleTranslator(source='auto', target='en').translate(x)
)
```

---

## ⚠️ Problem with this approach

* Slow for large datasets
* API called for every row
* Can freeze notebook

---

# 🚀 Best Practice (Professional Method)

## ✔ Step 1: Create translator once

```python
translator = GoogleTranslator(source='auto', target='en')
```

---

## ✔ Step 2: Safe function

```python
import pandas as pd

def translate_text(x):
    if pd.isnull(x):
        return x
    try:
        return translator.translate(x)
    except:
        return x
```

---

## ✔ Step 3: Apply function

```python
df["translated_text"] = df["text_column"].apply(translate_text)
```

---

# ⚡ Batch Translation (Advanced & Fast)

## ✔ Why batching?

* Faster processing
* Fewer API calls
* Less risk of freezing

---

## 🔧 Syntax

```python
texts = df["text_column"].fillna("").tolist()

translated = translator.translate_batch(texts)
```

---

## 📌 Example with batching function

```python
def batch_translate(text_list, batch_size=100):
    results = []
    
    for i in range(0, len(text_list), batch_size):
        batch = text_list[i:i+batch_size]
        
        translated_batch = translator.translate_batch(batch)
        results.extend(translated_batch)
        
        print(f"Processed batch {i} to {i+batch_size}")
    
    return results
```

---

# 📊 Real-World Use Case (Your Olist Project)

## Step 1: Translate reviews

```python
order_reviews["review_en"] = batch_translate(
    order_reviews["review_message"].fillna("").tolist(),
    batch_size=100
)
```

---

## Step 2: Sentiment Analysis (after translation)

```python
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

order_reviews["sentiment"] = order_reviews["review_en"].apply(
    lambda x: sia.polarity_scores(x)["compound"]
)
```

---

# 🧠 Key Concepts

## 🔹 Translation Flow

```
Original text → Language detection → Translation → Output text
```

---

## 🔹 Modes of usage

| Method            | Speed  | Best use        |
| ----------------- | ------ | --------------- |
| `.apply()`        | Slow   | small datasets  |
| Safe function     | Medium | medium datasets |
| Batch translation | Fast   | large datasets  |

---

# ⚠️ Common Mistakes

❌ Creating translator inside loop
❌ Translating full dataset without batching
❌ Not handling NaN values
❌ Not saving progress

---

# 💡 Best Practice Pipeline

```text
Raw reviews
   ↓
Clean missing values
   ↓
Batch translation (deep-translator)
   ↓
Sentiment analysis
   ↓
Customer segmentation
   ↓
Insights (churn analysis)
```

---

# 🔥 Final Summary

* `deep-translator` = Python translation library 🌍
* Best engine: `GoogleTranslator`
* Use `.translate()` for single text
* Use `.translate_batch()` for performance
* Always handle missing values
* Batch processing = professional approach 🚀

```

---
