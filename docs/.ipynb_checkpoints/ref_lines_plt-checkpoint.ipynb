# 📊 MATPLOTLIB HLINE & VLINE FUNCTIONS (FULL GUIDE)

---

# 🔹 1. `axhline()` (Axis Horizontal Line)

## 🧠 Simple definition:

Draws a **horizontal line across the entire plot** at a specific y-value.

---

## 🎯 What it is used for:

* Showing **mean/average lines**
* Showing **target/KPI thresholds**
* Highlighting important horizontal levels

---

## 🧾 Full syntax:

```python id="axh1"
plt.axhline(
    y,
    xmin=0,
    xmax=1,
    color=None,
    linestyle='-',
    linewidth=1,
    label=None
)
```

---

## 📌 Parameter explanation:

* **y** → position of the horizontal line (required)
* **xmin** → start position (0 = left edge)
* **xmax** → end position (1 = right edge, full width)
* **color** → line color
* **linestyle** → style of line (dashed, dotted, etc.)
* **linewidth** → thickness
* **label** → for legend

---

## 📍 Example:

```python id="axh2"
plt.axhline(y=50, color='red', linestyle='--', linewidth=2)
```

---

# 🔹 2. `hlines()` (Horizontal Line Segments)

## 🧠 Simple definition:

Draws a **horizontal line between specific x-ranges only** (not full width).

---

## 🎯 What it is used for:

* Highlighting **ranges or intervals**
* Marking specific sections of a chart
* Comparing segments

---

## 🧾 Full syntax:

```python id="hl1"
plt.hlines(
    y,
    xmin,
    xmax,
    colors=None,
    linestyles='solid',
    linewidths=1,
    label=None
)
```

---

## 📌 Example:

```python id="hl2"
plt.hlines(y=50, xmin=2, xmax=8, colors='green', linestyles='dotted')
```

---

# 🔹 3. `axvline()` (Axis Vertical Line)

## 🧠 Simple definition:

Draws a **vertical line across the entire plot** at a specific x-value.

---

## 🎯 What it is used for:

* Marking **events in time series**
* Highlighting thresholds on x-axis
* Separating sections in charts

---

## 🧾 Full syntax:

```python id="av1"
plt.axvline(
    x,
    ymin=0,
    ymax=1,
    color=None,
    linestyle='-',
    linewidth=1,
    label=None
)
```

---

## 📍 Example:

```python id="av2"
plt.axvline(x=5, color='blue')
```

---

# 🔹 4. `vlines()` (Vertical Line Segments)

## 🧠 Simple definition:

Draws a **vertical line between specific y-ranges only**.

---

## 🎯 What it is used for:

* Highlighting partial vertical ranges
* Marking segment boundaries
* Visual comparisons

---

## 🧾 Full syntax:

```python id="vl1"
plt.vlines(
    x,
    ymin,
    ymax,
    colors=None,
    linestyles='solid',
    linewidths=1,
    label=None
)
```

---

## 📍 Example:

```python id="vl2"
plt.vlines(x=5, ymin=2, ymax=10, colors='purple')
```

---

# 🔥 MULTIPLE `axhline()` PLOTTING

## 📌 Method 1: Direct multiple calls

```python id="mh1"
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4, 5], [10, 20, 15, 25, 30])

plt.axhline(y=10, color='red', linestyle='--')
plt.axhline(y=20, color='green', linestyle='--')
plt.axhline(y=25, color='blue', linestyle='--')

plt.show()
```

---

## 📌 Method 2: Using a loop (best practice)

```python id="mh2"
levels = [10, 20, 30, 40]

for y in levels:
    plt.axhline(y=y, linestyle='--', alpha=0.6)
```

---

## 🧠 Why loop is better:

* Cleaner code
* Easier scaling
* Works well with dynamic thresholds (KPIs, percentiles, etc.)

---

# 📊 QUICK SUMMARY TABLE

| Function  | Simple Meaning       | Full Coverage | Main Use             |
| --------- | -------------------- | ------------- | -------------------- |
| axhline() | Horizontal full line | Yes           | thresholds, averages |
| hlines()  | Horizontal segment   | No            | ranges, intervals    |
| axvline() | Vertical full line   | Yes           | events, markers      |
| vlines()  | Vertical segment     | No            | partial markers      |

---

# 🔥 FINAL KEY IDEA

* **axhline / axvline → full reference lines**
* **hlines / vlines → partial or segmented lines**
* multiple lines = just repeat function or loop
