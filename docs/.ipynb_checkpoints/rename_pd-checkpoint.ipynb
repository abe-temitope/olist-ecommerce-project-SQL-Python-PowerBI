# 🏷️ Pandas — Rename Columns Guide

---

# 🧠 Overview

Renaming columns in pandas helps you:
- Clean messy datasets
- Standardize naming for analysis
- Prepare data for tools like SQL, Tableau, Power BI

---

# 📊 1. `rename()` — MAIN METHOD

## 🔧 Full Syntax

```python
DataFrame.rename(
    mapper=None,
    index=None,
    columns=None,
    axis=None,
    inplace=False
)
````

---

## 🎯 Rename Specific Columns

```python
df = df.rename(columns={
    "old_name": "new_name",
    "price": "revenue"
})
```

---

## 💾 Using `inplace=True`

```python
df.rename(columns={"price": "revenue"}, inplace=True)
```

👉 Modifies the original DataFrame

---

## 🔄 Best Practice (Recommended)

```python
df = df.rename(columns={"price": "revenue"})
```

---

# 📊 2. Rename Multiple Columns

```python
df = df.rename(columns={
    "col1": "customer_id",
    "col2": "order_date",
    "col3": "total_price"
})
```

---

# 🧱 3. Rename Using Functions (POWERFUL 🔥)

## 🔹 Convert all column names to lowercase

```python
df = df.rename(columns=lambda x: x.lower())
```

---

## 🔹 Replace spaces with underscores

```python
df = df.rename(columns=lambda x: x.replace(" ", "_"))
```

---

## 🔹 Clean column names (Real-world usage)

```python
df = df.rename(columns=lambda x: x.strip().lower().replace(" ", "_"))
```

---

# 🚀 4. Direct Assignment (Alternative Method)

```python
df.columns = ["col1", "col2", "col3"]
```

⚠️ Rule:

```text
Number of new column names MUST match existing columns
```

---

# 📊 5. Rename Index (Rows)

```python
df = df.rename(index={
    0: "first_row",
    1: "second_row"
})
```

---

# 🎯 6. Using `axis`

```python
df = df.rename({"old_name": "new_name"}, axis=1)
```

👉 Same as `columns=`

---

# 🧠 REAL-WORLD EXAMPLES

## 🔥 Clean messy dataset

```python
df = df.rename(columns=lambda x: x.strip().lower().replace(" ", "_"))
```

---

## 🔥 Standardize for analytics tools

```python
df = df.rename(columns={
    "Customer ID": "customer_id",
    "Order Date": "order_date"
})
```

---

# ⚠️ COMMON MISTAKES

## ❌ Forgetting assignment

```python
df.rename(columns={"price": "revenue"})
```

👉 This does NOT change the DataFrame

---

## ❌ Wrong number of column names

```python
df.columns = ["a", "b"]  # ERROR if mismatch
```

---

# ⚡ QUICK CHEAT SHEET

| Task                | Code                           |
| ------------------- | ------------------------------ |
| Rename one column   | `df.rename(columns={"a":"b"})` |
| Rename multiple     | `df.rename(columns={...})`     |
| Clean all columns   | `lambda function`              |
| Replace all columns | `df.columns = [...]`           |

---

# 🧠 FINAL MINDSET

```text
rename() → safe and flexible
df.columns → full replacement
lambda → scalable cleaning
```

```

---

If you want next 🔥  
I can give you:
- **:contentReference[oaicite:0]{index=0}**
- OR a **:contentReference[oaicite:1]{index=1}** 📊
```
