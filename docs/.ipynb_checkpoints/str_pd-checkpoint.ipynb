# 🔤 Pandas String Functions (`.str`) — Full Guide

---

# 🧠 Overview

Pandas provides a `.str` accessor to apply **string operations on an entire column (Series)**.

```python
df["column"].str.<method>()
````

👉 Works only on **string (object) or string dtype columns**

---

# 📊 1. Case Conversion

## 🔹 Lowercase

```python
df["name"] = df["name"].str.lower()
```

## 🔹 Uppercase

```python
df["name"] = df["name"].str.upper()
```

## 🔹 Title Case

```python
df["name"] = df["name"].str.title()
```

---

# ✂️ 2. Remove Spaces

## 🔹 Strip (both sides)

```python
df["name"] = df["name"].str.strip()
```

## 🔹 Left strip

```python
df["name"] = df["name"].str.lstrip()
```

## 🔹 Right strip

```python
df["name"] = df["name"].str.rstrip()
```

---

# 🔍 3. Searching & Filtering

## 🔹 Contains

```python
df[df["name"].str.contains("john", case=False)]
```

---

## 🔹 Startswith

```python
df[df["name"].str.startswith("A")]
```

---

## 🔹 Endswith

```python
df[df["email"].str.endswith(".com")]
```

---

# 🔄 4. Replace Text

```python
df["name"] = df["name"].str.replace("old", "new")
```

---

# ✂️ 5. Split Strings

```python
df["first_name"] = df["full_name"].str.split(" ").str[0]
```

---

## 🔹 Expand into columns

```python
df[["first", "last"]] = df["full_name"].str.split(" ", expand=True)
```

---

# 🔗 6. Join Strings

```python
df["full"] = df["first"] + " " + df["last"]
```

---

# 🔢 7. Length of Strings

```python
df["length"] = df["name"].str.len()
```

---

# 🔍 8. Extract Patterns (Regex 🔥)

```python
df["numbers"] = df["text"].str.extract(r"(\d+)")
```

---

# 🔹 Extract multiple groups

```python
df[["area_code", "number"]] = df["phone"].str.extract(r"(\d{3})-(\d+)")
```

---

# 🔠 9. Check Data Type (String Checks)

```python
df["is_alpha"] = df["name"].str.isalpha()
df["is_digit"] = df["name"].str.isdigit()
df["is_lower"] = df["name"].str.islower()
```

---

# 🔄 10. Change Data Type to String

```python
df["col"] = df["col"].astype(str)
```

---

# ⚠️ IMPORTANT (Handling NaN)

Most `.str` methods fail with NaN unless handled:

```python
df["name"].str.lower()
```

### ✅ Safe way:

```python
df["name"].str.lower().fillna("")
```

OR

```python
df["name"].fillna("").str.lower()
```

---

# 🚀 REAL-WORLD CLEANING EXAMPLE

```python
df["name"] = (
    df["name"]
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

---

# ⚡ COMMONLY USED METHODS

| Method        | Purpose          |
| ------------- | ---------------- |
| `.lower()`    | lowercase text   |
| `.upper()`    | uppercase text   |
| `.strip()`    | remove spaces    |
| `.contains()` | search text      |
| `.replace()`  | replace values   |
| `.split()`    | split strings    |
| `.len()`      | string length    |
| `.extract()`  | regex extraction |

---

# 🧠 FINAL MINDSET

```text
.str → vectorized string operations (fast & scalable)
```

👉 Instead of looping:

```python
for x in df["col"]:
    ...
```

👉 Use:

```python
df["col"].str.method()
```

---

# 🎯 PRO TIP

```text
Chain multiple .str methods for powerful cleaning pipelines
```

Example:

```python
df["clean_name"] = (
    df["name"]
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```

---

If you want next 🔥  
I can give you:
- **:contentReference[oaicite:0]{index=0}**
- OR a **real messy dataset cleaning walkthrough using `.str` functions** 📊
```
