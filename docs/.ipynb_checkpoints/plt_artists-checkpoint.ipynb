# 🎨 Matplotlib Artists & Chart Types – Complete Guide

## 📌 Overview

In Matplotlib, **everything drawn on a plot is an Artist**. An Artist is any object that appears on a figure, including lines, bars, text, shapes, and legends.

This document explains:

* All major Artist types
* Which chart produces which Artist
* Syntax + examples
* How they are stored in `ax`
* A full visual cheatsheet

---

# 🧠 1. What is an Artist?

An **Artist** is a base object in Matplotlib that represents anything drawn on a figure.

## 🎯 Two main categories:

### 🧱 Primitive Artists

Basic visual elements:

* Line2D (lines)
* Text (labels)
* Patch (shapes)
* Collection (groups of points)

### 📦 Container Artists

Group multiple artists:

* BarContainer
* ErrorbarContainer
* Legend

---

# 📊 2. Line Charts (`ax.plot`)

## 🧾 Syntax

```python
ax.plot(x, y, label="Line A")
```

## 🎨 Artist Type

* `Line2D`

## 📍 Storage

```python
ax.lines
```

## 📊 Example

```python
fig, ax = plt.subplots()
ax.plot([1,2,3], [10,20,15], label="Sales")
```

## 🧠 Key Insight

Each line = one `Line2D` object

---

# 🔵 3. Scatter Plots (`ax.scatter`)

## 🧾 Syntax

```python
ax.scatter(x, y)
```

## 🎨 Artist Type

* `PathCollection`

## 📍 Storage

```python
ax.collections
```

## 📊 Example

```python
ax.scatter([1,2,3], [10,20,15])
```

## 🧠 Key Insight

All points grouped into one collection

---

# 🟦 4. Bar Charts (`ax.bar`)

## 🧾 Syntax

```python
ax.bar(x, height)
```

## 🎨 Artist Type

* `BarContainer`

## 📍 Storage

```python
ax.containers
```

## 📊 Example

```python
ax.bar(["A","B"], [10,20])
```

## 🧠 Key Insight

Each bar is inside a container

---

# 🟩 5. Horizontal Bar Charts (`ax.barh`)

## 🧾 Syntax

```python
ax.barh(y, width)
```

## 🎨 Artist Type

* `BarContainer`

## 📍 Storage

```python
ax.containers
```

---

# 🟠 6. Error Bar Charts (`ax.errorbar`)

## 🧾 Syntax

```python
ax.errorbar(x, y, yerr)
```

## 🎨 Artist Type

* `ErrorbarContainer`

## 📍 Storage

```python
ax.containers
```

## 📊 Example

```python
ax.errorbar([1,2,3], [10,20,15], yerr=[1,2,1])
```

---

# 🥧 7. Pie Charts (`ax.pie`)

## 🧾 Syntax

```python
ax.pie(values, labels=labels)
```

## 🎨 Artist Types

* Wedge (Patch)
* Text

## 📍 Storage

```python
ax.patches
ax.texts
```

## 🧠 Key Insight

Pie charts are NOT containers

# 🏷️ 10. Annotation (`ax.annotate`)

## 🧾 Syntax

```python
ax.annotate("Note", xy=(x,y))
```

## 🎨 Artist Types

* Text
* Optional Arrow (Patch)

## 📍 Storage

```python
ax.texts
ax.patches
```

---

# 📦 11. Legend (`ax.legend`)

## 🧾 Syntax

```python
ax.legend()
```

## 🎨 Artist Type

* Legend (Container-like object)

## 📍 Storage

```python
ax.legend_  # single legend object
```

---

# 🧠 12. FULL STORAGE MAP (VERY IMPORTANT)

```
ax
├── ax.lines        → Line2D (plot)
├── ax.collections  → Scatter, groups
├── ax.containers   → BarContainer, ErrorbarContainer
├── ax.patches      → Bars, pie slices, histogram blocks
├── ax.texts        → text + annotations
├── ax.legend_      → legend object
```

---

# 🧭 13. FULL VISUAL CHEATSHEET

```
CHART TYPE         ARTIST TYPE            STORAGE
--------------------------------------------------------
plot()            Line2D                ax.lines
scatter()         PathCollection        ax.collections
bar()             BarContainer          ax.containers
barh()            BarContainer          ax.containers
errorbar()        ErrorbarContainer     ax.containers
hist()            Patch                 ax.patches
pie()             Wedge + Text          ax.patches + ax.texts
text()            Text                  ax.texts
annotate()        Text + Patch          ax.texts + ax.patches
legend()          Legend                ax.legend_
```

---

# 🚀 14. Key Insights

* Everything = Artist
* Charts differ in Artist type
* Only bar-style plots use containers
* Other plots use lines, patches, or collections

---

# 💡 15. Mental Model

Think of Matplotlib like this:

🎭 Figure = Stage
🎨 Axes = Canvas
🧑‍🎨 Artists = Everything drawn

* Lines = strokes
* Bars = grouped blocks
* Scatter = point clouds
* Text = labels
* Patches = shapes
