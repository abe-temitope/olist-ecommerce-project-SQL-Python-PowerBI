# 🗑️ Pandas Drop Functions — Full Syntax Guide

---

# 📊 1. `drop()` — FULL SYNTAX

```python
DataFrame.drop(
    labels=None,
    axis=0,
    index=None,
    columns=None,
    level=None,
    inplace=False,
    errors='raise'
)
````

---

## 🔍 PARAMETERS EXPLAINED

### 🏷️ `labels`

* Row/column names to drop

```python
df.drop(labels=["A", "B"], axis=1)
```

---

### 🔄 `axis`

```text
0 → rows  
1 → columns
```

```python
df.drop("col1", axis=1)
df.drop(0, axis=0)
```

---

### 🎯 `index`

* Explicitly drop rows

```python
df.drop(index=[0, 2, 5])
```

---

### 📊 `columns` (BEST PRACTICE)

```python
df.drop(columns=["col1", "col2"])
```

---

### 🧱 `level` (MultiIndex only)

```python
df.drop("level_name", level=0)
```

---

### 💾 `inplace`

```python
df.drop(columns=["col1"], inplace=True)
```

---

### ⚠️ `errors`

```python
df.drop(columns=["col1"], errors="ignore")
```

* `"raise"` → default
* `"ignore"` → skips missing values

---

## ✅ REAL EXAMPLE

```python
df = df.drop(columns=["customer_id", "order_id"], errors="ignore")
```

---

# 📊 2. `dropna()` — FULL SYNTAX

```python
DataFrame.dropna(
    axis=0,
    how='any',
    thresh=None,
    subset=None,
    inplace=False
)
```

---

## 🔍 PARAMETERS

### 🔄 `axis`

```text
0 → rows  
1 → columns
```

---

### ❗ `how`

```text
'any' → drop if ANY null exists  
'all' → drop if ALL values are null  
```

```python
df.dropna(how="any")
df.dropna(how="all")
```

---

### 🎯 `thresh`

* Minimum number of NON-null values required

```python
df.dropna(thresh=3)
```

---

### 🎯 `subset`

* Check nulls in specific columns only

```python
df.dropna(subset=["price", "quantity"])
```

---

### 💾 `inplace`

```python
df.dropna(inplace=True)
```

---

## ✅ REAL EXAMPLE

```python
df = df.dropna(subset=["revenue"])
```

---

# 📊 3. `drop_duplicates()` — FULL SYNTAX

```python
DataFrame.drop_duplicates(
    subset=None,
    keep='first',
    inplace=False,
    ignore_index=False
)
```

---

## 🔍 PARAMETERS

### 🎯 `subset`

```python
df.drop_duplicates(subset=["customer_id"])
```

---

### 🔁 `keep`

```text
'first' → keep first occurrence  
'last' → keep last occurrence  
False → remove ALL duplicates  
```

```python
df.drop_duplicates(keep="last")
```

---

### 💾 `inplace`

```python
df.drop_duplicates(inplace=True)
```

---

### 🔢 `ignore_index`

```python
df.drop_duplicates(ignore_index=True)
```

---

## ✅ REAL EXAMPLE

```python
df = df.drop_duplicates(subset=["order_id"], keep="first")
```

---

# 📊 4. `pop()` — FULL SYNTAX

```python
DataFrame.pop(item)
```

---

## 🔍 PARAMETER

### 🎯 `item`

* Column name to remove

---

## ⚠️ IMPORTANT BEHAVIOR

```python
col = df.pop("revenue")
```

👉 This:

* Removes column from DataFrame
* Returns it as a Series

---

## ✅ REAL EXAMPLE

```python
revenue = df.pop("revenue")
```

---

# 🧠 FINAL COMPARISON

| Function            | Key Feature                    |
| ------------------- | ------------------------------ |
| `drop()`            | Most flexible (rows & columns) |
| `dropna()`          | Handles missing values         |
| `drop_duplicates()` | Cleans duplicate data          |
| `pop()`             | Removes + returns column       |

---

# 🎯 PRO TIP

```text
Use columns= instead of axis=1 → cleaner & safer
```

---

# 🚀 REAL-WORLD CLEANING FLOW

```python
# 1. Drop unnecessary columns
df = df.drop(columns=["unwanted_col"])

# 2. Remove missing values
df = df.dropna(subset=["important_col"])

# 3. Remove duplicates
df = df.drop_duplicates(subset=["id"])

# 4. Extract column if needed
target = df.pop("target_column")
```

---

# 🧠 FINAL MINDSET

```text
drop → structure  
dropna → quality  
drop_duplicates → integrity  
pop → extraction  
```

```

---

If you want next 🔥  
I can turn this into:
- :contentReference[oaicite:0]{index=0}
- OR :contentReference[oaicite:1]{index=1} 💼📊
```
