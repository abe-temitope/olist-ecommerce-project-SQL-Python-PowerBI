# Matplotlib `legend()` – Full Guide

## 🧾 Full Syntax

```python
ax.legend(handles=None, labels=None, loc=None, title=None,
          bbox_to_anchor=None, ncol=1, fontsize=None,
          frameon=True, shadow=False)
```

> You can also use `plt.legend()` if you're not working with `ax`.

---

## 🔍 Explanation of Key Parameters

### `handles`

* The plotted objects (lines, bars, etc.)
* Usually **automatic**, so you rarely need to pass this manually

---

### `labels`

* Names that appear in the legend
* Also usually **auto-generated** from `label="..."` in your plot

---

### `loc`

* Controls **where the legend appears**

**Common values:**

* `"best"` (default)
* `"upper right"`
* `"upper left"`
* `"lower left"`
* `"lower right"`
* `"center"`

---

### `title`

* Adds a **heading** to the legend

```python
title="Segment"
```

---

### `bbox_to_anchor`

* Fine-tunes **exact positioning**
* Uses coordinates `(x, y)`

  * `(0,0)` = bottom-left
  * `(1,1)` = top-right
* Often used to move legend **outside the plot**

---

### `ncol`

* Number of **columns** in the legend

```python
ncol=2
```

---

### `fontsize`

* Controls the size of legend text

```python
fontsize=10
```

---

### `frameon`

* Show/hide the legend box

```python
frameon=False
```

---

### `shadow`

* Adds a shadow behind the legend

```python
shadow=True
```

---

## 📊 Example (Full Usage)

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [15, 18, 22, 28]

fig, ax = plt.subplots()

ax.plot(x, y1, label="Product A")
ax.plot(x, y2, label="Product B")

ax.legend(
    title="Products",
    loc="upper left",
    bbox_to_anchor=(1.05, 1),
    fontsize=10,
    ncol=1,
    frameon=True,
    shadow=True
)

plt.show()
```

---

## 🧠 What This Example Does

* Plots two lines (**Product A & Product B**)
* Adds a legend with:

  * Title: **Products**
  * Positioned **outside the plot (right side)**
  * Styled with font size, frame, and shadow

---

## 💡 Pro Tip

Most of the time, you only need:
