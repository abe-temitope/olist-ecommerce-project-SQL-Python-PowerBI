## 🎨 Custom Color Palette Guide (E-commerce Project)

I’m standardizing my chart colors to keep all visuals consistent, clean, and professional across my analysis and dashboards.

This approach helps me:
- Maintain a consistent visual identity
- Improve readability of charts
- Highlight important insights clearly

---

## 🧠 Design Idea Behind My Palette

For this project, I’m using a color style inspired by modern e-commerce dashboards:

- A strong **primary color** to represent key metrics (like revenue)
- Supporting colors for comparisons and categories
- Clear alert colors for losses or issues
- Neutral tones to keep everything balanced and not overwhelming

---

## 🔥 My Color System

```python
COLORS = {
    "primary": "#FF6A00",     # Main focus (e.g., revenue)
    "secondary": "#2E86AB",   # Supporting comparisons
    "accent": "#D7263D",      # Alerts (returns, cancellations)
    "success": "#1B998B",     # Positive metrics (profit, growth)
    "neutral_dark": "#2F2F2F",
    "neutral_light": "#F5F5F5"
}

PALETTE = [
    "#FF6A00",
    "#2E86AB",
    "#1B998B",
    "#F6C85F",
    "#6F4E7C"
]
````

---

## ⚙️ Applying It Globally

I set this once so all my charts automatically follow the same style:

```python
import seaborn as sns

sns.set_theme(
    style="whitegrid",
    palette=PALETTE
)
```

---

## 📊 How I Use Colors in Charts

Instead of random colors, I assign meaning:

```python
# Revenue charts
color = COLORS["primary"]

# Profit / growth
color = COLORS["success"]

# Returns / cancellations
color = COLORS["accent"]
```

For charts with multiple categories, I use the full palette.

---
