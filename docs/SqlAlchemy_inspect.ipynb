# 📘 SQLAlchemy Inspector + Table Merging Cheat Sheet

---

## 🔌 Setup

```python
from sqlalchemy import create_engine, inspect

engine = create_engine("your_connection_string")
inspector = inspect(engine)
````

---

## 📂 Get Database Structure

### Get all tables

```python
inspector.get_table_names()
```

### Get all views

```python
inspector.get_view_names()
```

### Get all schemas

```python
inspector.get_schema_names()
```

---

## 📊 Get Table Information

### Get columns in a table

```python
inspector.get_columns("table_name")
```

### Example

```python
for col in inspector.get_columns("orders"):
    print(col["name"], col["type"])
```

---

### Get primary key

```python
inspector.get_pk_constraint("table_name")
```

---

### Get foreign keys

```python
inspector.get_foreign_keys("table_name")
```

---

### Get indexes

```python
inspector.get_indexes("table_name")
```

---

### Get view SQL definition

```python
inspector.get_view_definition("view_name")
```

---

## 🔁 Loop Through All Tables

```python
for table in inspector.get_table_names():
    print(f"\nTable: {table}")
    
    for col in inspector.get_columns(table):
        print(f"{col['name']} - {col['type']}")
```

---

## ⚠️ Common Error

### ❌ Error

```
NameError: inspect is not defined
```

### ✅ Fix

```python
from sqlalchemy import inspect
```

---

## 🧠 SQL Equivalent

| Task     | SQL                |
| -------- | ------------------ |
| Tables   | SHOW TABLES        |
| Views    | SHOW VIEWS         |
| Columns  | DESCRIBE table     |
| Metadata | INFORMATION_SCHEMA |

---
