# 📊 Pandas `rank()` — Full Guide

---

# 🧠 What is `rank()`?

```python
Series.rank()
DataFrame.rank()
````

👉 Assigns a **ranking (position)** to values in a column.

---

# 🎯 Core Idea

```text
Converts numeric values → ordered positions
```

* Smallest or largest gets rank **1** (depends on settings)

---

# 📌 BASIC EXAMPLE

```python
import pandas as pd

df = pd.DataFrame({
    "revenue": [100, 200, 200, 50]
})

df["rank"] = df["revenue"].rank()
print(df)
```

### Output:

| revenue | rank |
| ------- | ---- |
| 100     | 2.0  |
| 200     | 3.5  |
| 200     | 3.5  |
| 50      | 1.0  |

---

# 🔍 Why 3.5?

```text
Tied values share the average rank
(3 + 4) / 2 = 3.5
```

---

# ⚙️ FULL SYNTAX

```python
rank(
    axis=0,
    method='average',
    numeric_only=False,
    na_option='keep',
    ascending=True,
    pct=False
)
```

---

# 🔑 PARAMETERS EXPLAINED

---

## 🔁 1. `ascending`

```python
df["rank"] = df["revenue"].rank(ascending=False)
```

👉 Highest value gets **rank 1**

---

## 🎯 2. `method` (Handling Ties 🔥)

| Method      | Behavior               |
| ----------- | ---------------------- |
| `"average"` | average rank (default) |
| `"min"`     | lowest rank            |
| `"max"`     | highest rank           |
| `"first"`   | order of appearance    |
| `"dense"`   | no gaps                |

---

### Example:

```python
df["rank"] = df["revenue"].rank(method="dense", ascending=False)
```

### Output:

| revenue | rank |
| ------- | ---- |
| 200     | 1    |
| 200     | 1    |
| 100     | 2    |
| 50      | 3    |

---

## 📊 3. `pct` (Percentage Ranking)

```python
df["rank_pct"] = df["revenue"].rank(pct=True)
```

👉 Converts rank into **percentage (0 → 1)**

---

## ⚠️ 4. `na_option`

```text
keep → leave NaN as is  
top → treat NaN as highest  
bottom → treat NaN as lowest  
```

---

## 🔄 5. `axis`

```text
0 → rank down rows (default)  
1 → rank across columns  
```

---

# 🚀 REAL-WORLD USE (RFM ANALYSIS)

```python
rfm["recency_rank"] = rfm["recency_days"].rank(ascending=True)
rfm["frequency_rank"] = rfm["frequency"].rank(ascending=False)
rfm["monetary_rank"] = rfm["monetary"].rank(ascending=False)
```

---

# 💡 PRO TIP

```text
Use method="dense" for business scoring systems
```

---

# ⚡ QUICK USE CASES

| Goal               | Code                    |
| ------------------ | ----------------------- |
| Highest = 1        | `rank(ascending=False)` |
| Lowest = 1         | `rank(ascending=True)`  |
| No gaps ranking    | `method="dense"`        |
| Percentile ranking | `pct=True`              |

---

# 🧠 FINAL MINDSET

```text
rank() → transforms raw numbers into ordered positions
```

👉 Helps answer:

* Who is top?
* Who is bottom?
* Who is average?

```

---
