# 📊 `qcut()` in Pandas

## 📌 Definition

`qcut()` is a function in **Pandas** used to divide data into **equal-sized groups (quantiles)**.

> It ensures that each group contains approximately the same number of observations.

---

## 🧾 Syntax

```python
pd.qcut(x, q, labels=None)
````

---

## 📦 Parameters

| Parameter | Description                                              |
| --------- | -------------------------------------------------------- |
| `x`       | Input data (Series or array-like)                        |
| `q`       | Number of quantiles (e.g., 4 = quartiles, 5 = quintiles) |
| `labels`  | Optional labels for the resulting bins                   |

---

## 🔍 Basic Example

```python
import pandas as pd

df = pd.DataFrame({
    "sales": [10, 20, 30, 40, 50, 60, 70, 80]
})

df["quartile"] = pd.qcut(df["sales"], q=4)

print(df)
```

---

## 🏷️ Example with Labels

```python
df["quartile"] = pd.qcut(
    df["sales"],
    q=4,
    labels=["Low", "Medium", "High", "Very High"]
)
```

---

## 📊 How It Works

* Splits data into **quantiles**
* Each bin contains **equal number of data points**
* Bin ranges are automatically calculated

---

## 🆚 `qcut()` vs `cut()`

| Feature      | `qcut()`               | `cut()`              |
| ------------ | ---------------------- | -------------------- |
| Bin type     | Equal-sized groups     | Fixed intervals      |
| Distribution | Balanced               | May be uneven        |
| Use case     | Ranking & segmentation | Range-based grouping |

---

## 🛒 RFM Use Case Example

```python
rfm["R_score"] = pd.qcut(rfm["recency"], 5, labels=[5,4,3,2,1])
rfm["F_score"] = pd.qcut(rfm["frequency"], 5, labels=[1,2,3,4,5])
rfm["M_score"] = pd.qcut(rfm["monetary"], 5, labels=[1,2,3,4,5])
```

---

## ⚠️ Common Issue

### Duplicate Values Error

If your data has many duplicate values, `qcut()` may fail.

### ✅ Solution

```python
pd.qcut(data, q=4, duplicates="drop")
```

---

## 🚀 Use Cases

* Customer segmentation (RFM analysis)
* Ranking data into tiers
* Creating balanced groups
* Data binning for analysis

---

## 💡 Key Takeaway

* `qcut()` divides data into **equal-sized groups**
* Best for **ranking and segmentation tasks**
* Widely used in **customer analytics and business intelligence**

```
```
