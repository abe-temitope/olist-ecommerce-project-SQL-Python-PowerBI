# 🧠 🔑 MUST-KNOW PANDAS PARAMETERS (GLOBAL CONCEPTS)

---

# 💾 1. `inplace`

```text id="inplace_core"
inplace=True → modify original DataFrame  
inplace=False → return new DataFrame (default)
```

### Example:

```python id="x1"
df.drop(columns=["col"], inplace=True)
```

### Meaning:

* `True` → no need to reassign
* `False` → you must do:

```python id="x2"
df = df.drop(columns=["col"])
```

---

# ⚠️ 2. `errors`

```text id="errors_core"
controls what happens when pandas meets a problem
```

### Common values:

* `"raise"` → stop and throw error (default)
* `"ignore"` → skip error silently

```python id="x3"
df.drop(columns=["missing_col"], errors="ignore")
```

---

# 🎯 3. `axis`

```text id="axis_core"
axis defines direction of operation
```

| Value | Meaning |
| ----- | ------- |
| 0     | rows    |
| 1     | columns |

```python id="x4"
df.drop("col", axis=1)
```

---

# 📌 4. `subset`

```text id="subset_core"
apply operation only on selected columns
```

```python id="x5"
df.dropna(subset=["price", "revenue"])
```

👉 Meaning:
Only check these columns for missing values

---

# 🔢 5. `how`

```text id="how_core"
controls condition logic (especially in dropna)
```

| Value | Meaning                          |
| ----- | -------------------------------- |
| "any" | drop if ANY null exists          |
| "all" | drop only if ALL values are null |

---

# 📊 6. `thresh`

```text id="thresh_core"
minimum number of valid values required
```

```python id="x6"
df.dropna(thresh=3)
```

👉 Keep rows with at least 3 non-null values

---

# 🧾 7. `keep`

```text id="keep_core"
used in drop_duplicates()
```

| Value   | Meaning               |
| ------- | --------------------- |
| "first" | keep first occurrence |
| "last"  | keep last occurrence  |
| False   | remove all duplicates |

```python id="x7"
df.drop_duplicates(keep="last")
```

---

# 🎯 8. `subset` (again, VERY IMPORTANT)

Used in:

* `dropna()`
* `drop_duplicates()`
* `merge()` (conceptually similar filtering idea)

👉 Means:

```text id="subset_meaning"
only consider these columns for the operation
```

---

# 💾 9. `ignore_index`

```text id="ignore_index"
reset index after operation
```

```python id="x8"
df.drop_duplicates(ignore_index=True)
```

---

# 🔗 10. `on`, `left_on`, `right_on` (MERGE CONTEXT)

```text id="merge_core"
used to define join keys
```

```python id="x9"
pd.merge(df1, df2, on="id")
```

OR

```python id="x10"
pd.merge(df1, df2, left_on="id1", right_on="id2")
```

---

# 🧱 11. `suffixes`

```text id="suffix_core"
handles column name conflicts after merge
```

```python id="x11"
pd.merge(df1, df2, on="id", suffixes=("_left", "_right"))
```

---

# 🔁 12. `sort`

```text id="sort_core"
sort results after merge/groupby
```

```python id="x12"
pd.merge(df1, df2, on="id", sort=True)
```
