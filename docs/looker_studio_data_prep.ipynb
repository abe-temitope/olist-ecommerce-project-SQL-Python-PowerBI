```python
import pandas as pd
from sqlalchemy import create_engine, inspect
```

```python
username = "root"
password = "abelytics"
database = "olist_ecommerce_db"
host = "127.0.0.1:3306"
```

```python
engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}/{database}")
connection = engine.connect()
print("Connected successfully!")
```

```python
inspector = inspect(engine)
tables = inspector.get_table_names()
print(tables)
```

```python
orders = pd.read_sql("SELECT * FROM orders", engine)
order_items = pd.read_sql("SELECT * FROM order_items", engine)
order_payments = pd.read_sql("SELECT * FROM order_payments", engine)
order_reviews = pd.read_sql("SELECT * FROM order_reviews", engine)
products = pd.read_sql("SELECT * FROM products", engine)
sellers = pd.read_sql("SELECT * FROM sellers", engine)
customers = pd.read_sql("SELECT * FROM customers", engine)
geolocation = pd.read_sql("SELECT * FROM geolocation", engine)
pcnt = pd.read_sql("SELECT * FROM product_category_name_translation", engine)
```

```python
order_1 = pd.merge(orders, order_items, on="order_id", how="inner")
```

```python
order_2 = pd.merge(order_1, order_payments, on="order_id")
```

```python
customer_order = pd.merge(order_2, customers, on="customer_id")
```

```python
customer_order_seller = pd.merge(customer_order, sellers, on="seller_id")
```

```python
customer_order_seller_product = pd.merge(customer_order_seller, products, on="product_id")
```

```python
customer_order_seller_product_geo = pd.merge(
    customer_order_seller_product,
    geolocation,
    left_on="seller_zip_code_prefix",
    right_on="geolocation_zip_code_prefix"
)
```
