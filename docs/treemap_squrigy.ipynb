# 📦 Treemap using Your Style (fig, ax)

A **treemap** is used to visualize hierarchical or proportional data using rectangles. Larger values take more space, making it great for comparing categories.

Unlike bar or pie charts, treemaps are not built into Matplotlib directly, so we use an external library like **:contentReference[oaicite:0]{index=0}** or **:contentReference[oaicite:1]{index=1}** with Matplotlib.

---

# 🧾 Basic Syntax (Your Style)

```python
fig, ax = plt.subplots()

squarify.plot(
    sizes=values,
    label=labels,
    ax=ax
)

plt.axis("off")
plt.show()
````

---

# 🧠 Breakdown (Your Style Thinking)

## 📊 1. `fig`

👉 The full canvas (figure)

* Acts as the container for everything
* Like a dashboard page

---

## 📍 2. `ax`

👉 The plotting area

* This is where the treemap is drawn
* All rectangles are rendered inside `ax`

---

## 🟩 3. `squarify.plot()`

👉 This creates the treemap layout

* Converts values into rectangles
* Bigger values → bigger blocks

---

# 📦 Full Syntax Explained

```python id="tree1"
squarify.plot(
    sizes,     # numeric values (block sizes)
    label=None, # category names
    color=None, # colors for blocks
    alpha=1.0,  # transparency
    ax=None     # matplotlib axis
)
```

---

# 🔍 Example (Your Style)

```python id="tree2"
import matplotlib.pyplot as plt
import squarify

labels = ["Electronics", "Fashion", "Home"]
values = [500, 300, 200]

fig, ax = plt.subplots()

squarify.plot(
    sizes=values,
    label=labels,
    ax=ax
)

plt.axis("off")
plt.show()
```

---

# 🎨 Improved Example (Professional Style)

```python id="tree3"
import matplotlib.pyplot as plt
import squarify

labels = ["Electronics", "Fashion", "Home"]
values = [500, 300, 200]

colors = ["#FF6A00", "#2E86AB", "#1B998B"]

fig, ax = plt.subplots()

squarify.plot(
    sizes=values,
    label=labels,
    color=colors,
    alpha=0.8,
    ax=ax
)

ax.set_title("Revenue Distribution by Category")

plt.axis("off")
plt.show()
```

---

# 🧠 Parameter Explanation

## 📊 `sizes`

* Determines rectangle size
* Higher value = bigger block

---

## 🏷️ `label`

* Category names displayed inside blocks

---

## 🎨 `color`

* Custom colors for each block

---

## 🔥 `alpha`

* Controls transparency (0 → 1)

---

## 📍 `ax`

* Places treemap inside Matplotlib axis

---

# 📊 Real Project Use (Olist Style)

Treemaps are great for:

* Revenue distribution by product category
* Customer segmentation size
* Market share visualization
* Inventory distribution

---

# 🚀 Final Mental Model (Your Style)

```text id="tree_model"
fig → full canvas (dashboard)
ax  → drawing area
squarify.plot() → builds rectangles inside ax
```

---

# 💡 Key Takeaway

```python id="tree_final"
fig, ax = plt.subplots()
squarify.plot(...)
```

* `fig` = container
* `ax` = plotting area
* `squarify.plot()` = creates the treemap structure

---

```
```
