In **pandas**, “duplicate functions” usually refer to tools used to **detect, view, and remove repeated rows or values** in a dataset 📊

Let’s break it down clearly 👇

---

## 🔁 1. `duplicated()` — Detect duplicates

This function checks whether a row is a duplicate of a previous one.

### 📌 Syntax:

```python
df.duplicated()
```

### 📊 What it returns:

* `True` → row is a duplicate
* `False` → row is unique (first occurrence is kept)

### 🧪 Example:

```python
import pandas as pd

data = {
    "name": ["A", "B", "A", "C", "B"],
    "score": [10, 20, 10, 30, 20]
}

df = pd.DataFrame(data)

df.duplicated()
```

---

### ⚙️ Important parameters:

```python
df.duplicated(subset=None, keep='first')
```

| Parameter      | Meaning                                    |
| -------------- | ------------------------------------------ |
| `subset`       | Check duplicates based on specific columns |
| `keep='first'` | Marks first occurrence as False            |
| `keep='last'`  | Marks last occurrence as False             |
| `keep=False`   | Marks ALL duplicates as True               |

---

### 📌 Example (subset):

```python
df.duplicated(subset=["name"])
```

👉 Checks duplicates only in `name` column

---

## 🧹 2. `drop_duplicates()` — Remove duplicates

This removes repeated rows.

### 📌 Syntax:

```python
df.drop_duplicates()
```

---

### 🧪 Example:

```python
df_clean = df.drop_duplicates()
```

---

### ⚙️ Parameters:

```python
df.drop_duplicates(subset=None, keep='first', inplace=False)
```

| Parameter      | Meaning                     |
| -------------- | --------------------------- |
| `subset`       | Which columns to check      |
| `keep='first'` | Keeps first occurrence      |
| `keep='last'`  | Keeps last occurrence       |
| `keep=False`   | Removes ALL duplicates      |
| `inplace=True` | Modifies original DataFrame |

---

### 📌 Example (remove all duplicates):

```python
df.drop_duplicates(keep=False)
```

---

## 📊 3. Real-world use case

In data cleaning:

### Example:

* Customer data
* Orders table
* Survey responses

You often do:

```python
# Step 1: check duplicates
df.duplicated().sum()

# Step 2: remove duplicates
df = df.drop_duplicates()
```

---

## 🧠 Key difference

| Function            | Purpose             |
| ------------------- | ------------------- |
| `duplicated()`      | Identify duplicates |
| `drop_duplicates()` | Remove duplicates   |

---

## 🚀 Pro tip (Data Analyst level)

Before removing duplicates:

```python
df[df.duplicated(keep=False)]
```

👉 This shows ALL duplicate rows so you can inspect them first
