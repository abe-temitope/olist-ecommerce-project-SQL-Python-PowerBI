```python
from ydata_profiling import ProfileReport
import pandas as pd
from sqlalchemy import create_engine
```

```python
username = "root"
password = "abelytics"
database = "olist_ecommerce_db"
host = "127.0.0.1:3306"
```

```python
engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}/{database}")
connection = engine.connect()
print("Connected successfully!")
```

```python
tables = [
    "customers",
    "sellers",
    "products",
    "orders",
    "order_items",
    "order_payments",
    "order_reviews",
    "geolocation",
    "product_category_name_translation"
]
```

```python
db = {}

for table in tables:
    print(f"Loading {table}...")
    db[table] = pd.read_sql(f"SELECT * FROM {table}", engine)
```

```python
for table in db:
    report = ProfileReport(
            db[table],
            title=f"{table}_table_profiling",
            dataset={
                "description": f"This is a {table} profiling report table of the Olist Ecommmerce Dataset.",
                "author": "Abe Temitope Moses",
                "copyright_holder": "Bethrass Solution Limited.",
                "copyright_year": 2026
            }
        )
    # report.to_notebook_iframe()
    report.to_file(f"../reports/data_profiling/{table}_profiling.html")
```
