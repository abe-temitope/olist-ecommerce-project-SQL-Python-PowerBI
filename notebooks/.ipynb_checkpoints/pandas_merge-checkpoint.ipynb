# 📘 Pandas merge() – Complete Cheat Sheet

---

## 🧠 What is merge()?

`pd.merge()` is used to **combine two DataFrames based on common columns (keys)**.

👉 Equivalent to **SQL JOIN**

---

## 🔌 Basic Syntax

```python
pd.merge(left_df, right_df, on=None, how='inner')
````

---

## 🧩 Parameters Explained

| Parameter | Meaning                         |
| --------- | ------------------------------- |
| left_df   | First DataFrame                 |
| right_df  | Second DataFrame                |
| on        | Column(s) common to both tables |
| how       | Type of join                    |
| left_on   | Column from left table          |
| right_on  | Column from right table         |
| suffixes  | Handle duplicate column names   |

---

# 🔥 BASIC EXAMPLE

```python
import pandas as pd

customers = pd.DataFrame({
    "customer_id": [1, 2, 3],
    "name": ["Abe", "John", "Mary"]
})

orders = pd.DataFrame({
    "order_id": [101, 102, 103],
    "customer_id": [1, 2, 1],
    "price": [200, 150, 300]
})

merged = pd.merge(customers, orders, on="customer_id")
print(merged)
```

---

## ✅ Output

```
   customer_id   name  order_id  price
0            1    Abe       101    200
1            1    Abe       103    300
2            2   John       102    150
```

---

# 🔥 JOIN TYPES (VERY IMPORTANT)

---

## 1️⃣ INNER JOIN (default)

```python
pd.merge(a, b, on="id", how="inner")
```

👉 Keeps only matching rows

---

## 2️⃣ LEFT JOIN

```python
pd.merge(a, b, on="id", how="left")
```

👉 Keeps all rows from left table

---

## 3️⃣ RIGHT JOIN

```python
pd.merge(a, b, on="id", how="right")
```

👉 Keeps all rows from right table

---

## 4️⃣ FULL OUTER JOIN

```python
pd.merge(a, b, on="id", how="outer")
```

👉 Keeps everything (fills missing with NaN)

---

# 🔥 USING DIFFERENT COLUMN NAMES

```python
pd.merge(
    orders,
    customers,
    left_on="cust_id",
    right_on="customer_id"
)
```

---

# 🔥 MERGING ON MULTIPLE COLUMNS

```python
pd.merge(
    df1,
    df2,
    on=["customer_id", "order_id"]
)
```

---

# 🔥 SELECTING COLUMNS (BEST PRACTICE)

## Before merge (recommended)

```python
merged = pd.merge(
    orders[["order_id", "customer_id", "price"]],
    customers[["customer_id", "name"]],
    on="customer_id"
)
```

---

## After merge

```python
merged = merged[["order_id", "customer_id", "name", "price"]]
```

---

# 🔥 HANDLE DUPLICATE COLUMN NAMES

```python
pd.merge(
    df1,
    df2,
    on="id",
    suffixes=("_left", "_right")
)
```

---

# 🔥 ADD INDICATOR COLUMN

```python
pd.merge(a, b, on="id", how="outer", indicator=True)
```

## Output example:

```
_id   _merge
1     both
2     left_only
3     right_only
```

---

# 🔥 VALIDATE MERGE (VERY IMPORTANT FOR DATA QUALITY)

```python
pd.merge(a, b, on="id", validate="one_to_one")
```

## Options:

| Option       | Meaning               |
| ------------ | --------------------- |
| one_to_one   | No duplicates in both |
| one_to_many  | Left unique           |
| many_to_one  | Right unique          |
| many_to_many | No restriction        |

---

# 🔥 MERGE ON INDEX

```python
pd.merge(
    df1,
    df2,
    left_index=True,
    right_index=True
)
```

---

# 🔥 MULTIPLE MERGES

```python
merged = pd.merge(orders, order_items, on="order_id")
merged = pd.merge(merged, customers, on="customer_id")
```

---

# 🔥 REAL-WORLD PIPELINE

```python
merged_orders = pd.merge(
    orders,
    order_items,
    on="order_id"
)

customer_summary = (
    pd.merge(merged_orders, customers, on="customer_id")
    .groupby("customer_unique_id")
    .agg(
        total_orders=("order_id", "count"),
        total_revenue=("price", "sum")
    )
    .reset_index()
)
```

---

# 🧠 SQL vs Pandas

| SQL       | Pandas      |
| --------- | ----------- |
| SELECT    | df[["col"]] |
| JOIN      | pd.merge()  |
| ON        | on="col"    |
| LEFT JOIN | how="left"  |
| GROUP BY  | groupby()   |

---

# ⚠️ COMMON MISTAKES

❌ Forgetting `on`

```python
pd.merge(a, b)  # error or wrong result
```

❌ Duplicate columns confusion
👉 Use `suffixes`

❌ Wrong join type
👉 Always think: LEFT vs INNER

---

# 🚀 PRO TIPS

✅ Always reduce columns before merge

```python
orders = orders[["order_id", "customer_id", "price"]]
```

✅ Check duplicates before merging

```python
orders["customer_id"].duplicated().sum()
```

✅ Use `validate` to avoid silent errors

---

# 🎯 FINAL MINDSET

```text
SQL → JOIN tables
Pandas → merge DataFrames
```

👉 Same logic, different syntax

```

---

If you want next 🔥  
I can give you **merge vs join vs concat (the confusion killer guide)** — that’s where most analysts struggle.
```
